```python
import pandas as pd
import numpy as np

def fill_missing_only(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # 1) 텍스트형: 빈 문자열로 채움
    for c in ["medical_history", "family_medical_history"]:
        if c in out.columns:
            out[c] = out[c].fillna("")

    # 2) 범주형: edu_level → 'Unknown'
    if "edu_level" in out.columns:
        out["edu_level"] = out["edu_level"].fillna("Unknown")

    # 3) mean_working: gender×age 분위수 그룹 중앙값 → 전역 중앙값
    if {"mean_working", "gender", "age"}.issubset(out.columns):
        # age를 분위수 구간으로 (중복 구간 방지)
        age_bins = pd.qcut(out["age"], q=5, duplicates="drop")
        grp_key = pd.concat([out["gender"], age_bins], axis=1)

        # 각 행별 그룹 중앙값 벡터
        grp_median = (
            out[["mean_working"]]
            .assign(gender=grp_key.iloc[:,0].values, age_bin=grp_key.iloc[:,1].values)
            .groupby(["gender", "age_bin"])["mean_working"]
            .transform("median")
        )
        out["mean_working"] = out["mean_working"].fillna(grp_median)

        # 아직 남은 NaN은 전체 중앙값으로
        out["mean_working"] = out["mean_working"].fillna(out["mean_working"].median())

    # 4) 숫자 컬럼: 남아있는 NaN만 각 컬럼 중앙값으로
    num_cols = out.select_dtypes(include=[np.number]).columns
    for c in num_cols:
        if out[c].isna().any():
            out[c] = out[c].fillna(out[c].median())

    # 5) 기타 문자 컬럼: 남아있는 NaN만 'Unknown'으로
    obj_cols = out.select_dtypes(include=["object", "string"]).columns
    for c in obj_cols:
        if c not in ["medical_history", "family_medical_history", "edu_level"] and out[c].isna().any():
            out[c] = out[c].fillna("Unknown")

    return out
```

```python
class AddMissingFlags(BaseEstimator, TransformerMixin):
    def __init__(self, cols): self.cols = cols
    def fit(self, X, y=None): return self
    def transform(self, X):
        X = X.copy()
        for c in self.cols:
            if c in X.columns:
                X[c + "_missing"] = X[c].isna().astype(int)
        return X

class AgeBinner(BaseEstimator, TransformerMixin):
    def __init__(self, q=5, col="age", out_col="age_bin"):
        self.q, self.col, self.out_col = q, col, out_col
    def fit(self, X, y=None):
        s = pd.Series(X[self.col])
        self.edges_ = s.quantile(q=np.linspace(0,1,self.q+1), interpolation="linear").to_numpy()
        self.edges_[0], self.edges_[-1] = -np.inf, np.inf
        return self
    def transform(self, X):
        X = X.copy()
        X[self.out_col] = pd.cut(X[self.col], bins=self.edges_, include_lowest=True).astype(str)
        return X

class GroupMedianImputer(BaseEstimator, TransformerMixin):
    """gender×age_bin 그룹 중앙값으로 target_col 보간, 없으면 전역 중앙값"""
    def __init__(self, target_col="mean_working", group_cols=("gender","age_bin")):
        self.target_col = target_col
        self.group_cols = group_cols   # ← __init__에서 원형 그대로 저장(변환 금지)

    def fit(self, X, y=None):
        Xc = X.copy()
        # fit에서만 작업용 리스트 생성
        self.group_cols_ = list(self.group_cols)
        self.grp_median_ = (
            Xc.groupby(self.group_cols_, observed=True)[self.target_col].median()
        )
        self.global_median_ = Xc[self.target_col].median()
        return self

    def transform(self, X):
        X = X.copy()
        m = X[self.target_col].isna()
        if m.any():
            # 멀티인덱스 Series에서 안전하게 조회
            X.loc[m, self.target_col] = X.loc[m, self.group_cols_].apply(
                lambda r: self.grp_median_.get(tuple(r), np.nan), axis=1
            )
            X[self.target_col] = X[self.target_col].fillna(self.global_median_)
        return X

class OutlierFlagsIQR(BaseEstimator, TransformerMixin):
    """모든 숫자 컬럼에 IQR 이상치 플래그 추가"""
    def __init__(self, k=1.5): self.k = k
    def fit(self, X, y=None):
        self.num_cols_ = X.select_dtypes(include=[np.number]).columns.tolist()
        self.bounds_ = {}
        for c in self.num_cols_:
            s = pd.to_numeric(X[c], errors="coerce").dropna()
            if s.empty: continue
            q1, q3 = s.quantile([0.25, 0.75]); iqr = q3 - q1
            self.bounds_[c] = (q1 - self.k*iqr, q3 + self.k*iqr)
        return self
    def transform(self, X):
        X = X.copy()
        for c, (lo, hi) in self.bounds_.items():
            s = pd.to_numeric(X[c], errors="coerce")
            X[f"{c}_out_low"]  = ((s < lo) & s.notna()).astype(int)
            X[f"{c}_out_high"] = ((s > hi) & s.notna()).astype(int)
        return X

class WinsorizeColumn(BaseEstimator, TransformerMixin):
    """특정 컬럼을 IQR 펜스로 절삭(clip)"""
    def __init__(self, col="mean_working", k=1.5): self.col, self.k = col, k
    def fit(self, X, y=None):
        s = pd.to_numeric(X[self.col], errors="coerce").dropna()
        q1, q3 = s.quantile([0.25, 0.75]); iqr = q3 - q1
        self.lo_, self.hi_ = q1 - self.k*iqr, q3 + self.k*iqr
        return self
    def transform(self, X):
        X = X.copy()
        X[self.col] = pd.to_numeric(X[self.col], errors="coerce").clip(lower=self.lo_, upper=self.hi_)
        return X
```

```python
# build_pipeline() 전체 교체
def build_pipeline():
    num_base = [
        "age","height","weight","cholesterol",
        "systolic_blood_pressure","diastolic_blood_pressure",
        "glucose","bone_density","mean_working"
    ]
    cat_cols  = ["gender","activity","smoke_status","sleep_pattern","edu_level","age_bin"]
    flag_cols = ["edu_level","mean_working","medical_history","family_medical_history"]

    # 숫자 피처 + 플래그들
    num_cols = num_base + \
               [f"{c}_missing" for c in flag_cols] + \
               [f"{c}_out_low" for c in num_base] + \
               [f"{c}_out_high" for c in num_base]

    # 숫자/범주 변환기
    num_tf = Pipeline([("imp", SimpleImputer(strategy="median"))])
    cat_tf = Pipeline([
        ("imp", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore"))
    ])

    # 텍스트 변환기 (각각 별도 인스턴스 + flatten 추가)
    text_tf_mh = Pipeline([
        ("imp", SimpleImputer(strategy="constant", fill_value="")),
        ("flatten", FunctionTransformer(lambda X: X.astype(str).ravel(), validate=False)),
        ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=5, max_features=4000)),
    ])
    text_tf_fmh = Pipeline([
        ("imp", SimpleImputer(strategy="constant", fill_value="")),
        ("flatten", FunctionTransformer(lambda X: X.astype(str).ravel(), validate=False)),
        ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=5, max_features=4000)),
    ])

    # ColumnTransformer: 단일 컬럼도 반드시 리스트로!
    preproc = ColumnTransformer(
        transformers=[
            ("num", num_tf, num_cols),
            ("cat", cat_tf, cat_cols),
            ("mh",  text_tf_mh,  ["medical_history"]),
            ("fmh", text_tf_fmh, ["family_medical_history"]),
        ],
        remainder="drop",
        sparse_threshold=0.3
    )

    model = LGBMRegressor(
        n_estimators=1200, learning_rate=0.03,
        num_leaves=63, subsample=0.8, colsample_bytree=0.8,
        reg_alpha=0.1, reg_lambda=1.0, random_state=42
    )

    pipe = Pipeline([
        ("add_flags", AddMissingFlags(flag_cols)),
        ("age_bin", AgeBinner(q=5, col="age", out_col="age_bin")),
        ("grp_imp", GroupMedianImputer("mean_working", ("gender","age_bin"))),  # tuple 유지
        ("out_flags", OutlierFlagsIQR(k=1.5)),
        ("winsor_mw", WinsorizeColumn(col="mean_working", k=1.5)),
        ("preproc", preproc),
        ("model", model),
    ])
    return pipe

```

```python
from sklearn.model_selection import KFold, cross_validate
import numpy as np

target = "stress_score"
drop_cols = [c for c in ["ID", target] if c in df.columns]
X = df.drop(columns=drop_cols)
y = df[target].astype(float)

pipe = build_pipeline()

cv = KFold(n_splits=15, shuffle=True, random_state=42)
scoring = {"rmse": "neg_root_mean_squared_error", "mae": "neg_mean_absolute_error"}
cv_res = cross_validate(pipe, X, y, cv=cv, scoring=scoring, n_jobs=-1, return_train_score=False)

rmse = -cv_res["test_rmse"]; mae = -cv_res["test_mae"]

print("=== 교차검증(5-Fold) 결과 ===")
for i, (r, m) in enumerate(zip(rmse, mae), 1):
    print(f"Fold {i}: RMSE={r:.4f}, MAE={m:.4f}")
print(f"\n평균 RMSE: {rmse.mean():.4f} (표준편차 {rmse.std():.4f})")
print(f"평균 MAE : {mae.mean():.4f} (표준편차 {mae.std():.4f})")
```
